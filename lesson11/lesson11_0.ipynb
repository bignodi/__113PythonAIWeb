```python
## model_validate  驗證字串
## model_validate_json 驗證josn (一定是字串)

## 範例目的:
## 不需定義完整資料結構
## 資料類型 不是全採用str, 而是採用自訂義的類型並做檢查
## e.g. No 第512 筆是 No='T6500066'; wgs84aX/wgs84Y, 第76 & 170 為空值, 如不採str, 就有報錯
## 

import json
from pydantic import BaseModel, Field, field_validator
from pprint import pprint

class Record(BaseModel):
    公司名稱:str = Field(alias="Name_Ins")
    登記編號:str = Field(alias="No")
    地址:str = Field(alias="Address")
    統一編號:str = Field(alias="tax_ID_number")
    緯度:float = Field(alias="wgs84aX")
    經度:float = Field(alias="wgs84aY")

    ## 因經/緯度是float, records.75 & 169 為空字串, 故會報錯
    ## 用classmethod checkNullValue() 檢查空字串
    @field_validator("緯度","經度",mode='before')
    @classmethod
    def checkNullValue(cls, value):
        if value == '':
            return 0.0
        else:
            return value

class Result(BaseModel):
    records:list[Record]

## 用json.loads 載入資料
with open('新北市食品工廠清冊.json',encoding='utf-8') as file:
    json_content:str = file.read()
    all_data = json.loads(json_content)
    ## 用pprint 看all_data 的資料結構
    ## pprint(all_data)

result:Result = Result.model_validate(all_data['result'])

for rec in result.records:
    print(rec)
```
